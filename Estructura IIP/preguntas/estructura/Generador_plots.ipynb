```python
import pandas as pd
import re
import textwrap

import plotly.graph_objects as go
```

```python
estructura_2019=pd.read_excel('estructura 2019.xlsx',sheet_name='Maquina')
estructura_2021=pd.read_excel('estructura 2021.xlsx',sheet_name='Maquina')
estructura_2023=pd.read_excel('estructura 2023.xlsx',sheet_name='Maquina')
```

```python
def create_treemap_df(df):
    labels = []
    parents = []

    for _, row in df.iterrows():
        if row['Componente'] not in labels:
            labels.append(row['Componente'])
            parents.append(row['T'])
        
        if row['Variable'] not in labels:
            labels.append(row['Variable'])
            parents.append(row['Componente'])
        
        if row['Indicador'] not in labels:
            labels.append(row['Indicador'])
            parents.append(row['Variable'])
        
        #labels.append(row['Pregunta'])
        #parents.append(row['Indicador'])

    treemap_df = pd.DataFrame({'labels': labels, 'parents': parents})
    return treemap_df
```

```python
def extract_text_before_dot(text):
    match = re.match(r"([^\.]+)", text)  # Match all characters before the first dot
    return match.group(0) if match else text
```

```python
def wrap_text(text, width=30):
    return "<br>".join(textwrap.wrap(text, width=width))
```

```python
df_2019=create_treemap_df(estructura_2019)
df_2019['short_label'] = df_2019['labels'].apply(extract_text_before_dot)
df_2019['short_parent'] = df_2019['parents'].apply(extract_text_before_dot)

df_2021=create_treemap_df(estructura_2021)
df_2021['short_label'] = df_2021['labels'].apply(extract_text_before_dot)
df_2021['short_parent'] = df_2021['parents'].apply(extract_text_before_dot)

df_2023=create_treemap_df(estructura_2023)
df_2023['short_label'] = df_2023['labels'].apply(extract_text_before_dot)
df_2023['short_parent'] = df_2023['parents'].apply(extract_text_before_dot)
```

# Treemap

```python
wrapped_hovertext = [wrap_text(label, width=50) for label in df_2019.labels]

fig = go.Figure(go.Treemap(
    hovertext=wrapped_hovertext,
    hoverinfo="text",
    labels = df_2019.short_label,
    parents = df_2019.short_parent,
    sort=False,
    root_color="lightgrey"
))
fig.update_layout(
    margin = dict(t=0, l=0, r=0, b=0),
    )
fig.write_html('est_tree_2019.html')
fig
```
